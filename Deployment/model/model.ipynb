```python
import pandas as pd
import numpy as np
import pickle
```

```python
data = pd.read_csv('Salary_Data.csv')
```

```python
x = data['YearsExperience'].values.reshape(-1,1)
y = data['Salary'].values.reshape(-1,1)
```

```python
from sklearn.preprocessing import PolynomialFeatures
pf = PolynomialFeatures(degree=3,include_bias=False)
x = pf.fit_transform(x)
```

```python
from sklearn.linear_model import LinearRegression
Lr = LinearRegression()
```

```python
Lr.fit(x,y)
```

```python
pickle.dump(Lr,open('/home/vignesh/ML/Deployment/model.pkl','wb'))
```

```python
pickle.dump(pf,open('/home/vignesh/ML/Deployment/model_poly.pkl','wb'))
```

```python
import matplotlib.pyplot as plt
```

```python
s = np.linspace(1,15)
plt.rcParams.update({
    "lines.color": "white",
    "patch.edgecolor": "white",
    "text.color": "black",
    "axes.facecolor": "white",
    "axes.edgecolor": "lightgray",
    "axes.labelcolor": "white",
    "xtick.color": "white",
    "ytick.color": "white",
    "grid.color": "lightgray",
    "figure.facecolor": "black",
    "figure.edgecolor": "black",
    "savefig.facecolor": "black",
    "savefig.edgecolor": "black"})
plt.plot(data['YearsExperience'].values.reshape(-1,1),y)
plt.plot(s,Lr.predict(pf.fit_transform(s.reshape(-1,1))))
plt.ylabel('Salary')
plt.xlabel('Years Of Experience')
plt.savefig('/home/vignesh/ML/Deployment/static/graph.png')
```
